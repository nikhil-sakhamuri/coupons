### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes

    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination

    location of the user, destination, and the venue, and we mark the distance between each

    direction as the destination.
    - Weather: sunny, rainy, or snowy

3. Coupon attributes

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
bar_data = data.query("coupon == 'Bar'")
```

2. What proportion of bar coupons were accepted?


```python
bar_data['Y'].sum() / len(bar_data['Y'])
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
bar_data['Bar'].unique()
```

```python
bar_3_or_fewer = ['never', 'less1', '1~3']
bar_more_than_3 = ['4~8', 'gt8']

bar_3_or_fewer_data = bar_data.query(f'Bar in {bar_3_or_fewer}')
p1 = bar_3_or_fewer_data['Y'].sum() / len(bar_3_or_fewer_data['Y'])
print(p1)

bar_more_than_3_data = bar_data.query(f'Bar in {bar_more_than_3}')
p2 = bar_more_than_3_data['Y'].sum() / len(bar_more_than_3_data['Y'])
print(p2)

# plot p1 and p2
sns.barplot(x=['3 or fewer times', 'more than 3 times'], y=[p1, p2])
# add x and y labels
plt.xlabel('Number of times bar visited per month')
plt.ylabel('Proportion of bar coupons accepted')
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
bar_data['age'].unique()
```

```python
bar_freqs = ['1~3', '4~8', 'gt8']
ages = ['46', '26', '31', '41', '50plus', '36']

#get row ids of targeted drivers
row_ids = bar_data['Bar'].isin(bar_freqs) & bar_data['age'].isin(ages)

#filter for targeted drivers
df = bar_data[row_ids]
p1 = df['Y'].sum() / len(df['Y'])

#filter for everyone else
df2 = bar_data[~row_ids]
p2 = df2['Y'].sum() / len(df2['Y'])


#plot p1 and p2
sns.barplot(x=['Bar more than once a month and age over 25', 'Everyone else'], y=[p1, p2])
# add x and y labels
plt.xlabel('Status')
plt.ylabel('Proportion of bar coupons accepted')
# rotate x
plt.xticks(rotation=90)
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
#filter for targeted groups and get Y probability
df = bar_data.query(f'Bar in {bar_freqs} and passanger != "Kid(s)" and occupation != "Farming Fishing & Forestry"')
p1 = df['Y'].sum() / len(df['Y'])
print(p1)

#filter for everyone else and get Y probability
df2 = bar_data.query(f'~(Bar in {bar_freqs} and passanger != "Kid(s)" and occupation != "Farming Fishing & Forestry")')
p2 = df2['Y'].sum() / len(df2['Y'])

#plot p1 and p2
sns.barplot(x=['Bar > 1x Month, Passengers != Kid, Occupation != Farming, etc.', 'Everyone else'], y=[p1, p2])
# add x and y labels
plt.xlabel('Status')
plt.ylabel('Proportion of bar coupons accepted')
# rotate x
plt.xticks(rotation=90)
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
print(bar_data['passanger'].unique())
print(bar_data['maritalStatus'].unique())
print(bar_data['age'].unique())
print(bar_data['Bar'].unique())
print(bar_data['RestaurantLessThan20'].unique())
print(bar_data['income'].unique())
```

```python
#set up the lists for filtering
bar_freqs = ['1~3', 'gt8', '4~8']
under_30 = ['21', '26', 'below21']
restaurants_more_than_4 = ['4~8', 'gt8']
less_than_50k = ['$12500 - $24999', '$37500 - $49999', '$25000 - $37499', 'Less than $12500']

#filter for the first condition and get the Y probability
df1 = bar_data.query(f'Bar in {bar_freqs} and passanger != "Kid(s)" and maritalStatus != "Widowed"')
p1 = df1['Y'].sum() / len(df1['Y'])

#filter for the 2nd condition and get the Y probability
df2 = bar_data.query(f'Bar in {bar_freqs} and age in {under_30}')
p2 = df2['Y'].sum() / len(df2['Y'])

#filter for the 3rd condition and get the Y probability
df3 = bar_data.query(f'RestaurantLessThan20 in {restaurants_more_than_4} and income in {less_than_50k}')
p3 = df3['Y'].sum() / len(df3['Y'])

#plot p1, p2, and p3
sns.barplot(x=['Bar > 1x Month, Passengers != Kid, Not Widowed',
              'Bar > 1x Month and Under 30', 'Cheap Restaurants > 4x Month and Less than 50k Income'], y=[p1, p2, p3])
# add x and y labels
plt.xlabel('Status')
plt.ylabel('Proportion of bar coupons accepted')
# rotate x
plt.xticks(rotation=90)
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

Based on these observations it seems like the drivers that accepted the bar coupons were people who frequented bars more than once a month to begin with. This seems to hold whether looking at people over 25 and people under 30. On the other hand, it seems like people without a lot of disposable income (people who went to cheap restaurants > 4x a month and less than 50k income) were unlikely to accept the coupon. Addtionally, it seems like people without kids were inclined to accept the coupon as well. 

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

```python
coffee_data = data.query("coupon == 'Coffee House'")
coffee_data
```

```python
#first lets look at the overall proportions of people who accepted the coupon or not 
sns.countplot(coffee_data['Y'])
print(coffee_data['Y'].sum() / len(coffee_data))
```

```python
#now lets look at people under the age of 21 vs all other ages
under_21 = ['below21']

#filter for people 21 and under
df = coffee_data.query(f'age in {under_21}')
p1 = df['Y'].sum() / len(df)

#now get all other people
df2 = coffee_data.query(f'~(age in {under_21})')
p2 = df2['Y'].sum() / len(df2)

sns.barplot(x=['Under 21', 'Everyone else'], y = [p1, p2])
plt.xlabel('Age Group')
plt.ylabel('Proportion that Accepted Coupon')
```

```python
#now lets look at the spread between people who go to a coffee house 4+ times a week vs poeple who go less
coffee_4_plus = ['4~8', 'gt8']

df = coffee_data.query(f'CoffeeHouse in {coffee_4_plus}')
p1 = df['Y'].sum() / len(df)

df2 = coffee_data.query(f'CoffeeHouse not in {coffee_4_plus}')
p2 = df2['Y'].sum() / len(df2)

sns.barplot(x=['4+ times per Month', 'Less than 4 times per month'], y = [p1, p2])
plt.xlabel('Coffee House Frequency')
plt.ylabel('Proportion that Accepted Coupon')
```

```python
#now lets see if destination has an impact
destinations = coffee_data['destination'].unique()
p_values = []

for dest in destinations:
    df = coffee_data[coffee_data.destination == dest]
    p = df['Y'].sum() / len(df)
    p_values.append(p)
    
sns.barplot(destinations, p_values)
plt.xlabel('Destination')
plt.ylabel('Proportion of Accepted Coupons')
```

```python
#Now lets see if coupon expiration has any impact acceptance rate
expirations = coffee_data['expiration'].unique()
p_values = []

for exp in expirations:
    df = coffee_data[coffee_data.expiration == exp]
    p = df['Y'].sum() / len(df)
    p_values.append(p)
    
sns.barplot(expirations, p_values)
plt.xlabel('Expiration')
plt.ylabel('Proportion of Accepted Coupons')
```

```python
#Lets split up the incomes into buckets and see if there is any trend
coffee_data['income'].unique()

low_income = ['Less than $12500', '$12500 - $24999', '$25000 - $37499', '$37500 - $49999']
mid_income = ['$50000 - $62499', '$62500 - $74999', '$75000 - $87499', '$87500 - $99999']
high_income = ['$100000 or More']

incomes = [low_income, mid_income, high_income]
p_values = []

for inc in incomes:
    df = coffee_data.query(f'income in {inc}')
    print(len(df))
    p = df['Y'].sum() / len(df)
    p_values.append(p)
    

sns.barplot(['<$50k', '$50k - $100k', '>$100k'], p_values)
plt.xlabel('Income')
plt.ylabel('Proportion that Accepted Coupon')
```
